```python
import urllib.request
import requests

import http.client
import socket
import re

def is_website_online(host):
    """ This function checks to see if a host name has a DNS entry by checking
        for socket info. If the website gets something in return, 
        we know it's available to DNS.
    """
    try:
        socket.gethostbyname(host)
    except socket.gaierror:
        return False
    else:
        return True

def write_bs_to_file(websites):
    i = 0
    arr = []
    while i < len(websites):
        page = websites[i]
        if is_website_online(page):
                try:
                    if urllib.request.urlopen('http://' + page).getcode() == 200:
                        #print(page)
                        writing_to_text(page)
                        #print('yes')
                except:
                    print()
                else:
                    print()
        i += 1
    return arr

working_urls = write_bs_to_file(tester)
print(working_urls)
```

```python
#want to loop through all possible files in the folder, open them, and check for the encoding

def open_file(url):
    ans = ""
    path = '/Users/christopherbarnes/cmsc25910/Assignment8/Websites_to_Text_Files/' + url + ".txt"
    with open(path) as f:
        contents = f.readlines()
        
    for content in contents:
        ans += content
    return ans

print(open_file('combros.vn'))
```

```python
from bs4 import BeautifulSoup

def get_utf8_type(url):
    sample = open_file(url)
    soup = BeautifulSoup(sample, 'html.parser')
    soup.prettify()
    meta = soup.find_all('meta')
    #print(meta)
    
    for m in meta:
        string = str(m).lower()
        #print(string)
        if 'utf-8' in string:
            if 'http-equiv' in string:
                return 'long'
            elif 'meta charset' in string:
                return 'short'
    return 'none'
            

print(get_utf8_type('combros.vn'))
print(get_utf8_type('buzhi.com'))
print(get_utf8_type('instagram-press.com'))
print(get_utf8_type('id3.org'))
```

```python
from bs4 import BeautifulSoup

def content_language(url):
    sample = open_file(url)
    soup = BeautifulSoup(sample, 'html.parser')
    soup.prettify()
    string = str(soup)
    if 'lang=' in string:
        return 'prefered'
    elif 'http-equiv="content-language' in string:
        return 'not prefered'
    else:
        return 'not indicated'

    
    
#print(content_language('zacuto.com'))
print(content_language('eefoo.com'))
```
